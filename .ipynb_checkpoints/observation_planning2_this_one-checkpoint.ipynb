```python
def func_prior(row, lims):
# a function to craete a priority column for the data
    #lims=[1.9,2.5,2.8]
    if (row['mir']+row['emir']>pheo2['mir']-np.log10(lims[0]*5.))&\
    (row['radio']+row['eradio']>pheo2['radio']-np.log10(lims[0]*3.))&\
    (row['xray']+row['exray']>pheo2['xray']-np.log10(lims[0])):
        return 'A'
    elif (row['mir']+row['emir']>pheo2['mir']-np.log10(lims[1]*5.))&\
    (row['radio']+row['eradio']>pheo2['radio']-np.log10(lims[1]*3.))&\
    (row['xray']+row['exray']>pheo2['xray']-np.log10(lims[1])):
        return 'B'
    elif (row['mir']+row['emir']>pheo2['mir']-np.log10(lims[2]*5.))&\
    (row['radio']+row['eradio']>pheo2['radio']-np.log10(lims[2]*3.))&\
    (row['xray']+row['exray']>pheo2['xray']-np.log10(lims[2])):
        return 'C'
    else:
        return 'D'

#Algorithm to put object into the time list
# by putting all the object in the first criterion and find the best spot to put it. If that spot is already taken,
# the algorithm will keep going down the table until it found an avaliable spot. 
# caution: this could lead to some object to slide down the table such that it is impossible to object 
# (already set, or near the horizon)
    
# def put_obj(df, target, time):
#     count=0
#     for index, row in df.iterrows():
#         x=[np.abs(t-row['best_time']).value for t in time]
#         loc=np.where(x==np.min(x))[0][0]
#         if target[loc]==None:
#             count+=1
#             target[loc]=row['name']
#         else:
#             for i in list_al(len(time)*2):        
#                 if -1<loc+i<len(time):
#                     if target[loc+i]==None:
#                         count+=1
#                         target[loc+i]=row['name']
#                         break
#     return target    

def put_obj(df, target, time):
    count=0
    for index, row in df.iterrows():
        loc=np.argmin(np.abs(time-row['best_time']))
        if target[loc]==None:
            count+=1
#             target[loc]=row['name']
            target[loc]=row['chips']
        else:
            for i in list_al(len(time)*2):        
                if -1<loc+i<len(time):
                    if target[loc+i]==None:
                        count+=1
#                         target[loc+i]=row['name']
                        target[loc+i]=row['chips']
                        break
    return target  


def list_al(n):
# generating a list of [1,-1,2,-2,3,-3,...] where n is the total number of element
    a = np.empty((n,))
    a[::2] = 1
    a[1::2] = -1

    la=[-1]
    count,num=0,0
    while len(la)<len(a):
        if count<2:
            la.append(num)
            count+=1
        else:    
            count=0
            num+=1
    return (np.arange(n)-np.array(la))*a
```

```python
def exposure_redshift(row, flag='normal'):
    z=row['redshift']
    if flag=='normal':
        if z>=0.1:
            return 120.+(250.-120.)/(0.7-0.1)*(z-0.1)
        else:
            return 250.
    elif flag=='bad_seeing':
        return 250
```

### 2. Remove object and rerun the algorithm
    - 'find_min_air': we can change the start and end of time to create a new list of best observed time
    - 'create_time_list': take a new DataFrame (both from changing the best observed time and drop some objects) and create a new time list of observation

```python
midnight = Time('2016-7-4 00:00:00',precision=0) - utcoffset
delta_midnight = np.linspace(0, 7.5, 76)*u.hour
times = midnight + delta_midnight

def find_min_air(df):
    list_obj=[SkyCoord(np.array(df['RA0'])[i], np.array(df['DEC0'])[i], frame='icrs', unit='deg') \
              for i in range(len(df['RA0']))]

    seczs_min_l,seczs_argmin_l=[],[]
    for i in list_obj:
        x=i.transform_to(AltAz(obstime=midnight+delta_midnight, location=lco_location))
        seczs_min_l.append(np.min(abs(x.secz-1.0)).value+1.)
        seczs_argmin_l.append(np.argmin(abs(x.secz-1.0)))

    df.loc[:,'min_air']=seczs_min_l
    df.loc[:,'best_time']=map(lambda x: x, (times-4*u.hr)[seczs_argmin_l])
    return df
```

```python
def create_time_list(df):
    lists=[]
    target1=dict.fromkeys(np.arange(len(times)))
    for i,item in enumerate(['A','B','C','D']):
        x=df.groupby('priority',axis=0).get_group(item)
        x_ind=x.set_index(np.arange(x.shape[0]))
        lists.append(x_ind)
        target1=put_obj(x_ind,target1,times-4.*u.hour)
        
    df_data1=pd.DataFrame({'time':(times-4.*u.hour).value,'name':target1.values()})
    total1=df_data1.merge(df, on='name', how='left')
    total1=total1[['name','time','RA0','DEC0','redshift','min_air','best_time','priority']]

    ra1=np.array(total1['RA0'])
    dec1=np.array(total1['DEC0'])

    list_objt=[]
    for i in range(len(ra1)):
        list_objt.append(SkyCoord(ra1[i], dec1[i], frame='icrs', unit='deg'))

    seczs_l,alts_l,azs_l=[],[],[]
    for i, item in enumerate(list_objt):
        x=item.transform_to(AltAz(obstime=times[i], location=lco_location))
        seczs_l.append(x.secz.value)
        alts_l.append(x.alt.value)
        azs_l.append(x.az.value)

    total1.loc[:,'obs_air']=seczs_l
    total1.loc[:,'obs_alt']=alts_l
    total1.loc[:,'obs_az']=azs_l
    print 'Number of None in the list:', np.sum(np.isnan(total1['obs_air']))
    return total1
```
